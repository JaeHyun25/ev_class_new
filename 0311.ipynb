```python
# 1776개(전체 시계열) NaN 값인 컬럼 제거
filtered_df = df.loc[:, df.isna().sum()<1776]

# 각 컬럼별 NaN 개수 계산
nan_counts_filtered = filtered_df.isna().sum()
```

```python
nan_counts_filtered
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 한글 폰트 적용
plt.rc('font', family='NanumGothic')

# 마이너스(-) 기호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False
```

```python
# NaN 값이 있는 컬럼만 필터링
nan_counts_filtered = nan_counts_filtered[nan_counts_filtered > 0]

# 시각화 (NaN 개수가 많은 컬럼만 표시시)
plt.figure(figsize=(15,7))
nan_counts_filtered.sort_values(ascending=False)[:50].plot(kind="bar")
plt.xlabel('센서')
plt.ylabel('NaN 값 개수')
plt.title('NaN 값 분포(Top 50)')
plt.xticks(rotation=90)
plt.show()
```

```python
nan_counts_filtered.value_counts()
```

103개의 NaN 값을 가진 센서가 982개란 뜻...\

```python
# nan 값을 가진 컬럼들의 시계열 데이터 분포 확인
nan_columns = nan_counts_filtered.index
nan_df = df[['timestamp'] + list(nan_columns)]
nan_df.head()
```

```python
nan_df.info()
```

```python
# 시간별 Nan 개수 집계
nan_time_series = nan_df.set_index('timestamp').isna().sum(axis=1)

# 시각화
plt.figure(figsize=(15,6))
plt.plot(nan_time_series.index, nan_time_series, marker='o', linestyle='-', alpha=0.6)
plt.xlabel('시간')
plt.ylabel('nan 개수')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# 타임스탬프 데이터 변환 (datetime 형식)
nan_df['timestamp'] = pd.to_datetime(nan_df['timestamp'])

# 시간과 일자 추출
nan_df["date"] = nan_df["timestamp"].dt.date
nan_df["hour"] = nan_df["timestamp"].dt.hour

# 일자별 24시간 동안의 NaN 개수 계산
nan_hourly = nan_df.groupby(["date", "hour"]).apply(lambda x : x.isna().sum().sum()).unstack(level=0)

# 시각화
plt.figure(figsize=(15, 8))
nan_hourly.plot(kind="line", marker="o", linestyle="-", alpha=0.7, colormap="viridis")
plt.xlabel("Hour of the Day")
plt.ylabel("Number of NaN Values")
plt.title("NaN Value Distribution by Hour (Grouped by Date)")
plt.legend(title="Date", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.xticks(range(0, 24))
plt.grid(True)
plt.show()
```

```python
# NaN 개수가 가장 많은 센서 TOP 을 날짜별로 분석
nan_top_per_day = nan_df.groupby("date").apply(lambda x: x.isna().sum().sort_values(ascending=False))
nan_top_per_day
```

```python
nan_top_per_day.value_counts()
```

```python
import matplotlib.pyplot as plt

# 날짜별 NaN 개수 합산
nan_per_day = nan_df.groupby("date").apply(lambda x: x.isna().sum().sum())

# 시각화
plt.figure(figsize=(12, 6))
plt.plot(nan_per_day.index, nan_per_day, marker="o", linestyle="-")
plt.xlabel("Date")
plt.ylabel("Total NaN Count")
plt.title("Total NaN Values per Day")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# NaN 발생이 가장 많은 센서 TOP 10 확인
nan_counts_per_sensor = nan_df.isna().sum().sort_values(ascending=False)
top_nan_sensors = nan_counts_per_sensor.head(10)

print(top_nan_sensors)
```

```python
# NaN이 포함된 데이터 비율 계산
nan_ratio = nan_df.isna().sum().sum() / (nan_df.shape[0] * nan_df.shape[1]) * 100

print(f"전체 데이터에서 NaN 비율: {nan_ratio:.2f}%")
```
